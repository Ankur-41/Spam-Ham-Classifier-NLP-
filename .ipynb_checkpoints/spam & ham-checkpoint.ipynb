```python
df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
df['spam/ham'] = df['spam/ham'].apply(lambda x : 1 if x == 'ham' else 0)
```

```python
import string

def remove_punc(text):
    return text.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_emojis(words): 
    new = ''
    for word in words:
        if word.isascii(): 
            new +=word
    return new
```

```python
df['text'] = df['text'].apply(remove_emojis)
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(txt): 
    words = word_tokenize(txt)
    words = [lemmatizer.lemmatize(word,pos='v')for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
df['text'] = df['text'].apply(preprocess)
```

```python
df.head()
```

### Splitting the dataset into independent and dependent feature

```python
x = df['text']
y = df['spam/ham']
```

### Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.256,random_state=12)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

bow_vectorizer = CountVectorizer()
tfidf_vectorizer = TfidfVectorizer()

x_train_bow = bow_vectorizer.fit_transform(x_train)
x_test_bow = bow_vectorizer.transform(x_test)

x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
x_test_tfidf = tfidf_vectorizer.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score

models = {
    'MultinomialNB' : MultinomialNB(),
    "LogisticRegression" : LogisticRegression(), 
    'SVC' : SVC()
}

def evaluate_score(true,pred,pred_type):
    acc_score = accuracy_score(true,pred)
    rec_score = recall_score(true,pred)
    pre_score = precision_score(true,pred)
    f_score = f1_score(true,pred)
    pred_type = pred_type
    return acc_score,rec_score,pre_score,f_score,pred_type
```

### BOW vectorization prediction

```python
mdl_lst = []
acc_lst = []
rec_lst = [] 
pre_lst = []
f1_lst = []
pred_lst = []

for name,model in models.items(): 
    model.fit(x_train_bow,y_train)
    y_train_pred_bow = model.predict(x_train_bow)
    y_test_pred_bow = model.predict(x_test_bow)

    train_acc_score,train_rec_score,train_pre_score,train_f_score,pred_type = evaluate_score(y_train,y_train_pred_bow,'Train')
    mdl_lst.append(name)
    acc_lst.append(train_acc_score)
    rec_lst.append(train_rec_score)
    pre_lst.append(train_pre_score)
    f1_lst.append(train_f_score)
    pred_lst.append(pred_type)
    
    test_acc_score,test_rec_score,test_pre_score,test_f_score,pred_type = evaluate_score(y_test,y_test_pred_bow,'Test')
    mdl_lst.append(name)
    acc_lst.append(test_acc_score)
    rec_lst.append(test_rec_score)
    pre_lst.append(test_pre_score)
    f1_lst.append(test_f_score)
    pred_lst.append(pred_type)

```

```python
pd.DataFrame(list(zip(mdl_lst,pred_lst,acc_lst,rec_lst,pre_lst,f1_lst)),columns=['Model (BOW)','Pred_type','Accuracy','Recall_score',
                                                                                'Precision_score','F1_score'])
```

### Tfidf Vectorization Prediction

```python
tf_mdl_lst = []
tf_acc_lst = []
tf_rec_lst = [] 
tf_pre_lst = []
tf_f1_lst = []
tf_pred_lst = []


for name,model in models.items(): 
    model.fit(x_train_tfidf,y_train)
    y_train_pred_tfidf = model.predict(x_train_tfidf)
    y_test_pred_tfidf = model.predict(x_test_tfidf)

    train_acc_score,train_rec_score,train_pre_score,train_f_score,pred_type = evaluate_score(y_train,y_train_pred_tfidf,'Train')
    tf_mdl_lst.append(name)
    tf_acc_lst.append(train_acc_score)
    tf_rec_lst.append(train_rec_score)
    tf_pre_lst.append(train_pre_score)
    tf_f1_lst.append(train_f_score)
    tf_pred_lst.append(pred_type)
    
    test_acc_score,test_rec_score,test_pre_score,test_f_score,pred_type = evaluate_score(y_test,y_test_pred_tfidf,'Test')
    tf_mdl_lst.append(name)
    tf_acc_lst.append(test_acc_score)
    tf_rec_lst.append(test_rec_score)
    tf_pre_lst.append(test_pre_score)
    tf_f1_lst.append(test_f_score)
    tf_pred_lst.append(pred_type)
```

```python
pd.DataFrame(list(zip(tf_mdl_lst,tf_pred_lst,tf_acc_lst,tf_rec_lst,tf_pre_lst,tf_f1_lst)),columns=['Model (TF-IDF)','Pred_type','Accuracy','Recall_score',
                                                                                'Precision_score','F1_score'])
```
